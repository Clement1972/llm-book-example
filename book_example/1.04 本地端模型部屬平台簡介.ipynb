# Open AI API 串接
Ollama 與 LM Studio 的模型部署都支援OpenAI API相容模式，因此兩者模型API的串接方式一樣。  
Ollama的預設port是11434，而LM Sudio的預測port則是1234。

## 文字補全

```python
# 從 LangChain 載入 OpenAI 類別，用於連接純文字補全的 API 伺服器
from langchain_openai import OpenAI 

model_name = 'gemma-3-12b-it'  # 指定模型名稱
base_url = 'http://localhost:1234/v1'  # 設定 LM Studio 本地伺服器的URL
# base_url = 'http://localhost:11434/v1'  # 設定 Ollama 本地伺服器的URL

# 建立 OpenAI 的實例，作為 LLM 的介面。
# model=model_name：指定模型的名稱（僅作標識）。
# openai_api_key="not-needed"：OpenAI 相容 API 雖要求此參數，但在 LM Studio 或 Ollama 中不需要真正的金鑰。
# openai_api_base=base_url：**最關鍵的設定**，將請求導向 LM Studio 或 Ollama 的本地伺服器。
llm = OpenAI(
    model=model_name, 
    openai_api_key="not-needed", # LM Studio 或 Ollama 不需金鑰
    openai_api_base=base_url    # 指向本地伺服器
)

# 呼叫（Invoke）LLM，詢問「機器學習的定義」。
# 此方法會發送請求給 LM Studio 或 Ollama ，並等待模型的文字回應。
result = llm.invoke("機器學習的定義")

# 變數 result 是一個包含模型生成文字的 Python 字串（string）物件。
print(result)
```

## 聊天補全

```python
# 從 langchain_openai 載入 ChatOpenAI 類別
from langchain_openai import ChatOpenAI

# 從 langchain_core.messages 載入 HumanMessage，用於構造聊天輸入
from langchain_core.messages import HumanMessage

model_name = 'gemma-3-12b-it'  # 指定模型名稱
base_url = 'http://localhost:1234/v1'  # 設定 LM Studio 本地伺服器的URL
# base_url = 'http://localhost:11434/v1'  # 設定 Ollama 本地伺服器的URL


# 建立 ChatOpenAI 的實例，作為聊天 LLM 的介面。
llm = ChatOpenAI(
    model=model_name,
    openai_api_key="not-needed", # LM Studio 或 Ollama 不需金鑰
    openai_api_base=base_url     # 指向本地伺服器
)

# 呼叫（Invoke）LLM，詢問「機器學習的定義」。
# [關鍵變更]：聊天模型需要傳入一個包含訊息物件（如 HumanMessage）的列表。
messages = [
    HumanMessage(content="機器學習的定義")
]

# 此方法會發送請求給 LM Studio 或 Ollama，並等待模型的 ChatGeneration 回應。
# result 是一個包含模型回應的 ChatMessage 物件 (通常是 AIMessage)。
result = llm.invoke(messages)

# [關鍵變更]：聊天模型的回應是一個訊息物件，需透過 .content 屬性來取得純文字。
print(result.content)
```
