# Document 物件

```python
from langchain_core.documents import Document
doc = Document("你好！")
doc
```

```python
doc = Document(
    page_content="你好！",
    metadata={"來源": "範例", "分類": "問候"}
)
doc
```

```python
doc = Document(
    page_content="今天會下雨嗎？",
    metadata={
        "來源": "新聞網站",
        "分類": "天氣",
        "日期": "2025-10-28",
        "重要性": "高"
    }
)
doc
```

```python
doc = Document(
    page_content="今天會下雨嗎？",
    metadata={"來源": "新聞網站", "分類": "天氣"},
    id="doc_001"
)
doc
```

# TXT DocumentLoader

```python
# from langchain.document_loaders import TextLoader
from langchain_community.document_loaders import TextLoader

# 建立 TXT Loader，指定檔案路徑
loader = TextLoader("data/CV.txt", encoding="utf-8")

# 載入文字檔並轉換成 Document 物件
documents = loader.load()
documents
```

# PDF DocumentLoader

```python
from langchain_community.document_loaders import PyPDFLoader

# 載入 PDF 文件
loader = PyPDFLoader("data/CV.pdf")
# 將 PDF 拆分成 Document（每一頁一個）
documents = loader.load()
documents
```

# Word DocumentLoader

```python
from langchain_community.document_loaders import UnstructuredWordDocumentLoader

# 使用 Loader 載入Word文件
loader = UnstructuredWordDocumentLoader("data/CV.docx")
documents = loader.load()
documents
```

# Excel DocumentLoader

```python
from langchain_community.document_loaders import UnstructuredExcelLoader
loader = UnstructuredExcelLoader("data/Iris.xlsx")
documents = loader.load()
documents
```

# CSV DocumentLoader

```python
from langchain_community.document_loaders import CSVLoader
loader = CSVLoader("data/Iris.csv", encoding="utf-8")
documents = loader.load()
documents
```

# HTML DocumentLoader

```python
from langchain_community.document_loaders import UnstructuredHTMLLoader
loader = UnstructuredHTMLLoader("data/CV.html")
documents = loader.load()
documents
```

# Web DocumentLoader

```python
from langchain_community.document_loaders import UnstructuredURLLoader
urls = ["https://sites.google.com/site/clement1972/"]
loader = UnstructuredURLLoader(urls=urls)
documents = loader.load()
documents
```
