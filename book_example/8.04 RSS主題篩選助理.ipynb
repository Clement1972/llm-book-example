## Gemini

```python
from langchain_google_genai import ChatGoogleGenerativeAI

API_KEY = "這邊請改成你自己的API_KEY值"
model_name = 'gemini-2.5-flash'

llm = ChatGoogleGenerativeAI(
    model=model_name,
    google_api_key=API_KEY,
    temperature=0
)
```

## LM Studio 或 Ollama

```python
from langchain_openai import ChatOpenAI
model_name = 'gemma-3-4b-it'  # 指定模型名稱，模型名稱會根據下載的模型不同而改變

# 根據使用LM Studio或Ollama來選擇適當的伺服器URL
base_url = 'http://localhost:1234/v1'  # LM Studio 本地伺服器的URL
# base_url = 'http://localhost:11434/v1' # Ollama 本地伺服器的URL

llm = ChatOpenAI(
    model=model_name,
    openai_api_key="not-needed",
    openai_api_base=base_url,
    temperature=0
)
```

## 測試模型

```python
from langchain_core.messages import HumanMessage
messages = [
    HumanMessage("機器學習的定義")
]

for chunk in llm.stream(messages):
    print(chunk.content, end='')
```

# 技術測試

## 測試RSS API

```python
import feedparser

# RSS 來源網址（可以換成你要的 RSS）
rss_url = "https://feeds.feedburner.com/techbang/daily"

# 解析 RSS
feed = feedparser.parse(rss_url)

# 顯示標題與摘要
print(f"Feed Title: {feed.feed.title}\n")
for entry in feed.entries[:5]:  # 只取前 5 筆
    print("標題:", entry.title)
    print("連結:", entry.link)
    print("日期:", entry.published if "published" in entry else "N/A")
    print("摘要:", entry.summary[:100], "...")
    print("-" * 50)
```

## Email Sender

```python
import smtplib
from email.mime.text import MIMEText

# Gmail SMTP 設定
smtp_server = "smtp.gmail.com"
port = 587
sender_email = "寄件者信箱"
password = "密碼" 

# 郵件內容
msg = MIMEText("這是一封測試郵件。", "plain", "utf-8")
msg["Subject"] = "Gmail SMTP 測試"
msg["From"] = sender_email
msg["To"] = "收件者信箱"

# 建立連線並寄信
with smtplib.SMTP(smtp_server, port) as server:
    server.starttls()
    server.login(sender_email, password)
    server.send_message(msg)

print("✅ 測試郵件寄送成功！")
```

## 使用LLM檢查主題類別

```python
from langchain_core.messages import HumanMessage, SystemMessage


def is_topic_relevant(topic, feed_entrie):
    system_prompt = f'''
    請分析使用者輸入的RSS訊息是否跟下面的主題有關？
    主題：{topic}
    
    輸出的內容為 'YES' 或 'NO'，不要進行解釋
    '''
    
    messages = [
            SystemMessage(content=system_prompt),
            HumanMessage(content=str(feed_entrie)[:1000]), # 這裡因為rss entrie可能太長，所以只限制前1000個字
        ]
    
    response = llm.invoke(messages)
    # print(response.content)
    return 'YES' in response.content
    
topic = ['手機', 'AI', '遊戲']
is_topic_relevant(topic, feed.entries[0]) # 測試答案
```

```python
topic = ['手機', 'AI', '遊戲']
for entry in feed.entries:
    response = is_topic_relevant(topic, entry)
    print(response, ": 標題:", entry.title)
```

# RSS主題篩選助理

```python
rss_list = [
    'https://www.digitimes.com.tw/tech/rss/xml/xmlrss_10_0.xml',
    'https://www.digitimes.com.tw/tech/rss/xml/xmlrss_10_10.xml',
    'https://www.digitimes.com.tw/tech/rss/xml/xmlrss_10_30.xml',
    'https://www.digitimes.com.tw/tech/rss/xml/xmlrss_10_40.xml',
    'https://www.digitimes.com.tw/tech/rss/xml/xmlrss_10_50.xml',
    'https://www.digitimes.com.tw/tech/rss/xml/xmlrss_10_60.xml',
    'https://www.digitimes.com.tw/tech/rss/xml/xmlrss_10_70.xml',
    'https://www.digitimes.com.tw/tech/rss/xml/xmlrss_10_90.xml',
    'https://www.digitimes.com.tw/tech/rss/xml/xmlrss_10_100.xml',
]
topic = ['手機', 'AI', '遊戲']
email_content = f'以下是符合主題:{topic}的文章：\n'

for rss in rss_list:
    try:
        feed = feedparser.parse(rss)
        print(f"\nFeed Title: {feed.feed.title}")
        email_content = email_content + f"Feed Title: {feed.feed.title}\n"        
        for entry in feed.entries:
            response = is_topic_relevant(topic, entry)
            print(response, ": 標題:", entry.title)
            if response:
                email_content = email_content + f'標題: {entry.title}, URL: {entry.link}\n'
        
    except:
        pass

# Gmail SMTP 設定
smtp_server = "smtp.gmail.com"
port = 587
sender_email = "寄件者信箱箱"
password = "密碼"

# 郵件內容
msg = MIMEText(email_content, "plain", "utf-8")
msg["Subject"] = "每日RSS列表"
msg["From"] = sender_email
msg["To"] = "收件者信箱"

# 建立連線並寄信
with smtplib.SMTP(smtp_server, port) as server:
    server.starttls()
    server.login(sender_email, password)
    server.send_message(msg)

print("✅ 測試郵件寄送成功！")
```
