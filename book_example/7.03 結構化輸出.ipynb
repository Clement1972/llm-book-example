# 選擇模型
請自由任意選擇下面兩個模型之一來進行範例  

## Gemini

```python
from langchain_google_genai import ChatGoogleGenerativeAI

API_KEY = "這邊請改成你自己的API_KEY值"
model_name = 'gemini-2.5-flash'

llm = ChatGoogleGenerativeAI(
    model=model_name,
    google_api_key=API_KEY,
    temperature=0
)
```

## LM Studio 或 Ollama

```python
from langchain_openai import ChatOpenAI
model_name = 'gemma-3-12b-it'  # 指定模型名稱，模型名稱會根據下載的模型不同而改變

# 根據使用LM Studio或Ollama來選擇適當的伺服器URL
base_url = 'http://localhost:1234/v1'  # LM Studio 本地伺服器的URL
# base_url = 'http://localhost:11434/v1' # Ollama 本地伺服器的URL

llm = ChatOpenAI(
    model=model_name,
    openai_api_key="not-needed",
    openai_api_base=base_url,
    temperature=0
)
```

## 測試模型

```python
from langchain_core.messages import HumanMessage
messages = [
    HumanMessage("機器學習的定義")
]

for chunk in llm.stream(messages):
    print(chunk.content, end='')
```

# CommaSeparatedListOutputParser

```python
from langchain_core.output_parsers import StrOutputParser

chain = llm | StrOutputParser()
```

```python
response = chain.invoke("列出五種熱帶水果的名字")
print(response)
```

```python
prompt = '''
請列出五種熱帶水果的名字
依照下面格式輸出
水果1, 水果2, 水果3, ...
'''
response = chain.invoke(prompt)
print(response)
```

```python
response.split(",")
```

```python
from langchain_core.output_parsers import CommaSeparatedListOutputParser
comma_parser = CommaSeparatedListOutputParser()
comma_parser.invoke(response)
```

```python
comma_chain = llm | CommaSeparatedListOutputParser()
```

```python
prompt = '''
請列出五種熱帶水果的名字
依照下面格式輸出
水果1, 水果2, 水果3, ...
'''
comma_chain.invoke(prompt)
```

```python
format_instructions = comma_parser.get_format_instructions()
format_instructions
```

```python
prompt = f'''
請列出五種熱帶水果的名字
{format_instructions}
'''
comma_chain.invoke(prompt)
```

```python
prompt = f'''
請列出五種熱帶水果的中文名字
{format_instructions}
'''
comma_chain.invoke(prompt)
```

# JsonOutputParser

```python
from langchain_core.output_parsers import JsonOutputParser
json_parser = JsonOutputParser()
```

```python
json_parser.get_format_instructions()
```

```python
json_chain = llm | json_parser
```

```python
prompt = f'''
請列出五種熱帶水果的中文名字
{json_parser.get_format_instructions()}
'''
json_chain.invoke(prompt)
```

```python
prompt = f'''
請列出五種熱帶水果
請輸出以下 JSON 結構：
- name: 水果中文名稱
- english_name: 水果英文名稱
- description: 水果描述
- origin: 水果產地

{json_parser.get_format_instructions()}
'''
json_chain.invoke(prompt)
```

# PydanticOutputParser

```python
from pydantic import BaseModel, Field
from typing import List

class TropicalFruit(BaseModel):
    name: str = Field(description="水果中文名稱")
    english_name: str = Field(description="水果英文名稱")
    description: str = Field(description="水果描述")
    origin: str = Field(description="水果產地")

class TropicalFruitList(BaseModel):
    fruits: List[TropicalFruit]
```

```python
from langchain_core.output_parsers import PydanticOutputParser

pydantic_parser = PydanticOutputParser(
    pydantic_object=TropicalFruitList
)
```

```python
pydantic_chain = llm | pydantic_parser
```

```python
prompt = f'''
請列出五種熱帶水果
{pydantic_parser.get_format_instructions()}
'''
response = pydantic_chain.invoke(prompt)
response
```

```python
prompt = f'''
請列出五種熱帶水果
{pydantic_parser.get_format_instructions()}
'''
response = json_chain.invoke(prompt)
response
```
