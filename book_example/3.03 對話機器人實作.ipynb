## Gemini

```python
from langchain_google_genai import ChatGoogleGenerativeAI

API_KEY = "這邊請改成你自己的API_KEY值"
model_name = 'gemini-2.5-flash'

llm = ChatGoogleGenerativeAI(
    model=model_name,
    google_api_key=API_KEY
)
```

## LM Studio 或 Ollama

```python
from langchain_openai import ChatOpenAI
model_name = 'gemma-3-12b-it'  # 指定模型名稱，模型名稱會根據下載的模型不同而改變

# 根據使用LM Studio或Ollama來選擇適當的伺服器URL
base_url = 'http://localhost:1234/v1'  # LM Studio 本地伺服器的URL
# base_url = 'http://localhost:11434/v1' # Ollama 本地伺服器的URL

llm = ChatOpenAI(
    model=model_name,
    openai_api_key="not-needed",
    openai_api_base=base_url 
)
```

## 測試模型

```python
from langchain_core.messages import HumanMessage
messages = [
    HumanMessage("機器學習的定義")
]
result = llm.invoke(messages)
print(result.content)
```

# Chat Bot

```python
from langchain_core.messages import HumanMessage, SystemMessage
class ChatBot:
    def __init__(self, llm):
        self.llm = llm #對話機器人的大腦        
        
        #對話機器人的記憶體
        self.messages = [SystemMessage("你是一個個人助理")] 

    # 接收使用者的對話內容
    def chat(self, text):        
        # 將使用者的對話內容新增到記憶中
        self.messages.append(HumanMessage(text))
        response = self.llm.invoke(self.messages) #將記憶內容交給LLM
        self.messages.append(response) #將LLM的回應內容加入記憶
        return response.content    

bot = ChatBot(llm) #創建一個對話機器人的實例
```

```python
msg = bot.chat("早安~~我是陳仁政！我喜歡讀書、寫程式")
print(msg)
```

```python
msg = bot.chat("今天天氣很好！")
print(msg)
```

```python
msg = bot.chat("我剛剛有做過自我介紹嗎？")
print(msg)
```

# 對話交談介面

```python
# 載入 Gradio 套件，用於建立網頁互動介面
import gradio as gr  

# ChatBot是我們先前定義好的對話機器人類別
# 以 llm (大型語言模型) 為參數建立一個機器人實例
bot = ChatBot(llm)  

# 定義主要的對話處理函式，供 Gradio 呼叫
def chat_function(message, history):
    '''
    用來處理每輪使用者輸入的對話內容。
    參數：
        message：使用者在這一輪輸入的訊息（字串）
        history：目前對話的歷史紀錄（list），包含過去所有輪的問答
    回傳：
        機器人根據輸入 message 產生的回覆內容（字串）
    '''
    # 呼叫我們的聊天機器人，取得回應
    response = bot.chat(message)
    # 將回應回傳給 Gradio，顯示在介面上
    return response

# 建立一個 Gradio 聊天介面
# - chat_function：指定每次用戶輸入時要執行的函式
webui = gr.ChatInterface(chat_function)

# 啟動 Gradio 網頁伺服器
# 預設會在本機開啟一個網頁 (例如 http://127.0.0.1:7860)
# 使用者可以透過瀏覽器與聊天機器人互動
webui.launch()
```
