## Gemini

```python
from langchain_google_genai import ChatGoogleGenerativeAI

API_KEY = "這邊請改成你自己的API_KEY值"
model_name = 'gemini-2.5-flash'

llm = ChatGoogleGenerativeAI(
    model=model_name,
    google_api_key=API_KEY,
    temperature=0
)
```

# Summary

```python
from langchain_core.chat_history import InMemoryChatMessageHistory

chat_history = InMemoryChatMessageHistory()

chat_history.add_user_message("影像物件偵測有？那些常見的方法")

response =  llm.invoke(chat_history.messages)
chat_history.add_message(response)

chat_history.add_user_message("YOLO也是影像物件偵測的方法嗎？")

response =  llm.invoke(chat_history.messages)
chat_history.add_message(response)
```

```python
for msg in chat_history.messages:
    print(f"[{msg.type}] {msg.content}\n")
```

```python
from langchain_core.prompts import ChatPromptTemplate

summary_prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一個對話摘要器，請用精簡但完整的方式更新摘要"),
    ("human", """
目前的對話摘要：
{summary}

新增的對話內容：
{new_lines}

請輸出更新後的摘要：
""")
])
```

```python
summary_chain = summary_prompt | llm
```

```python
def update_summary(summary: str, history: InMemoryChatMessageHistory) -> str:
    new_lines = "\n".join(
        f"{type(m).__name__}: {m.content}"
        for m in history.messages
    )

    result = summary_chain.invoke({
        "summary": summary,
        "new_lines": new_lines
    })

    return result.content
```

```python
summary = ""
summary = update_summary(summary, chat_history)
print(summary)
```

```python
total_chars = sum(len(m.content) for m in chat_history.messages)
print("chat_history 總字元數：", total_chars)
print("summary 總字元數：", len(summary))
```
