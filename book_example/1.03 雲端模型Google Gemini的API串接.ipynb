## 使用Gemini來生成文字
【提醒】
- 執行下面的程式可以會花數十秒的時間，因為需要等待Gemini生成全部文字後才會同時輸出內容。
- 後續的範例會示範如何使用串流的方式來即時顯示已生成的文字。

```python
# 從 LangChain 載入 GoogleGenerativeAI 類別，用於連接 Google Gemini 模型
from langchain_google_genai import GoogleGenerativeAI 

# 指定模型名稱
# 'gemini-2.5-flash' 是較快速且經濟的模型，適合大多數一般任務。
# 付費版（或需要較高能力時）則可考慮使用 'gemini-2.5-pro'。
model_name = 'gemini-2.5-flash'

# 建立 GoogleGenerativeAI 的實例，作為 LLM 的介面。
# model=model_name：指定要使用的 Gemini 模型。
# google_api_key=API_KEY：傳入您的 Google AI API 金鑰進行身份驗證。
llm = GoogleGenerativeAI(model=model_name, google_api_key=API_KEY)

# 呼叫（Invoke）LLM，詢問「機器學習的定義」。
# 此方法會發送請求給 Gemini 模型，並等待模型的文字回應。
result = llm.invoke("機器學習的定義")

# 變數 result 是一個包含模型生成文字的 Python 字串（string）物件。
print(result)
```
