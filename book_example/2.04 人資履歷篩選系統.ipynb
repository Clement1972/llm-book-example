## Gemini

````python
from langchain_google_genai import ChatGoogleGenerativeAI

API_KEY = "這邊請改成你自己的API_KEY值"
model_name = 'gemini-2.5-flash'

llm = ChatGoogleGenerativeAI(
    model=model_name,
    google_api_key=API_KEY
)
````

## LM Studio 或 Ollama

````python
from langchain_openai import ChatOpenAI
model_name = 'gemma-3-12b-it'  # 指定模型名稱，模型名稱會根據下載的模型不同而改變

# 根據使用LM Studio或Ollama來選擇適當的伺服器URL
base_url = 'http://localhost:1234/v1'  # LM Studio 本地伺服器的URL
# base_url = 'http://localhost:11434/v1' # Ollama 本地伺服器的URL

llm = ChatOpenAI(
    model=model_name,
    openai_api_key="not-needed",
    openai_api_base=base_url 
)
````

## 測試模型

````python
from langchain.messages import HumanMessage
messages = [
    HumanMessage("機器學習的定義")
]
result = llm.invoke(messages)
print(result.content)
````

## 封裝成一個 Python 函式
**需注意的是，LLM 生成的 JSON 資料可能存在格式不穩定的情況，存在一定機率導致 json.loads() 發生解析錯誤，建議加入例外處理或驗證機制。**


````python
from langchain.messages import HumanMessage, SystemMessage
import json
def ResumeAnalysis(llm, cv, job_content)-> dict:
    """
    依據個人履歷與工作需求，使用 LLM 進行適任性分析。
    
    Args:
        llm: 可執行 invoke 的語言模型實例 (LangChain Runnable)。
        cv (str): 求職者的完整個人履歷內容。
        job_content (str): 該職位的工作需求與條件描述。
        
    Returns:
        dict: 包含是否合適、適任度與詳細分析的 JSON 結果。
    """

    system_prompt = '''
    你是一個人資面試官，請依據下面的個人履歷跟工作需求來分析求職者是否適合這份工作：
    請嚴格按照下面的JSON格式來進行輸出：
    {"是否合適":"YES"或是"NO","適任度":0~100的評分,"分析":"詳細的理由"}
    
    請勿在輸出的 JSON 內容前後使用任何程式碼區塊標記（例如 ```json 或 ```）。
    '''
    
    user_prompt = f'''
    個人履歷：
    {cv}
    
    工作需求：
    {job_content}
    '''
    messages = [
        SystemMessage(system_prompt),
        HumanMessage(user_prompt)
    ]
    result = llm.invoke(messages)
    text = result.content
    start = text.find("{")
    end = text.rfind("}") + 1
    json_str = text[start:end]
    return json.loads(json_str)
````

````python
# 測試函式
ResumeAnalysis(llm, cv, job_content_01)
````
